```python
#Read data from excel files
#File location might vary


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

file = open('/Users/mjalife/Desktop/data_contour/fall0.csv')
fall0 = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall20s.csv')
fall20s = pd.read_csv(file)

file = open('/Users/mjalife/Desktop/data_contour/fall20n.csv')
fall20n = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall40s.csv')
fall40s = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall40n.csv')
fall40n = pd.read_csv(file)

file = open('/Users/mjalife/Desktop/data_contour/fall60s.csv')
fall60s = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall60n.csv')
fall60n = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall80s.csv')
fall80s = pd.read_csv(file)

file = open('/Users/mjalife/Desktop/data_contour/fall80n.csv')
fall80n = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall90s.csv')
fall90s = pd.read_csv(file)

file= open('/Users/mjalife/Desktop/data_contour/fall90n.csv')
fall90n = pd.read_csv(file)

fall0
```

```python
#Grab the density column for each latitude and put it into dataframe


data_fall = pd.DataFrame()
data_fall['height, km']= fall0['height, km']
lst = ['-90', '-80', '-60', '-40', '-20', '0', '20', '40', '60', '80', '90']
tmp = [fall90s, fall80s, fall60s, fall40s, fall20s, fall0, fall20n, fall40n, fall60n, fall80n, fall90n]
i = 0
for frame in lst:
    data_fall['{}'.format(frame)] = tmp[i]['Mass_density, g/cm-3']
    i = i+1

data_fall
```

```python
#Take the log of all the entires, remove the height column

array = data_fall.to_numpy()
test = np.delete(array, 0, axis=1)
test2 = np.log10(test)
test2
```

```python
#Polyfit, 2 for each latitude (one of degree a and one of degree b)

a = 6
b = 14

y = np.linspace(40,500, num=921)
x = test2[:,1]
p1 = np.polyfit(x, y, a)
p11 = np.polyfit(x, y, b)

y2 = np.linspace(40,500, num=921)
x2 = test2[:,3]
p2= np.polyfit(x2, y2, a)
p22= np.polyfit(x2, y2, b)

y3 = np.linspace(40,500, num=921)
x3 = test2[:,5]
p3 = np.polyfit(x3, y3, a)
p33 = np.polyfit(x3, y3, b)

y4 = np.linspace(40,500, num=921)
x4 = test2[:,7]
p4 = np.polyfit(x4, y4, a)
p44 = np.polyfit(x4, y4, b)

y5 = np.linspace(40,500, num=921)
x5 = test2[:,9]
p5 = np.polyfit(x5, y5, a)
p55 = np.polyfit(x5, y5, b)


```

```python
#make figure, change font size and figure size

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(ncols = 2, nrows=3)

plt.rcParams.update({'font.size': 22})
fig.set_figheight(18)
fig.set_figwidth(22)


#Three lines on each ax, one for data, and two for the polyfit

data80s = ax1.plot(test2[:,1], y, color = 'red')
regression80s = ax1.plot(x, np.polyval(p1, x), linestyle='dashed', color = 'blue')
regression80s1 = ax1.plot(x, np.polyval(p11, x), linestyle='dashed', color = 'green')

data40s = ax2.plot(test2[:,3], y2, color = 'red')
regression40s = ax2.plot(x2, np.polyval(p2, x2), linestyle='dashed', color = 'blue')
regression40s2 = ax2.plot(x2, np.polyval(p22, x2), linestyle='dashed', color = 'green')

data0 = ax3.plot(test2[:,5], y3, color = 'red')
regression0 = ax3.plot(x3, np.polyval(p3, x3), linestyle='dashed', color = 'blue')
regression03 = ax3.plot(x3, np.polyval(p33, x3), linestyle='dashed', color = 'green')

data40n = ax4.plot(test2[:,7], y4, color = 'red')
regression40n = ax4.plot(x4, np.polyval(p4, x4), linestyle='dashed', color = 'blue')
regression40n4 = ax4.plot(x4, np.polyval(p44, x4), linestyle='dashed', color = 'green')

data80n = ax5.plot(test2[:,9], y5, color = 'red')
regression80n = ax5.plot(x5, np.polyval(p5, x5), linestyle='dashed', color = 'blue')
regression80n5 = ax5.plot(x5, np.polyval(p55, x5), linestyle='dashed', color = 'green')



#Formatting axis

ax1.set_ylim(40, 150)
ax1.set_xlim(-12, -5)
ax1.set_ylabel('height / km')
ax1.set_xlabel('density')
ax1.set_title('North Hemisphere fall 80 S')
ax1.set_yticks([40,60,80,100, 120, 140])
ax1.grid(True)


ax2.set_ylim(40, 150)
ax2.set_xlim(-12, -5)
ax2.set_ylabel('height / km')
ax2.set_xlabel('density')
ax2.set_title('North Hemisphere fall 40 S')
ax2.set_yticks([40,60,80,100, 120, 140])
ax2.grid(True)

ax3.set_ylim(40, 150)
ax3.set_xlim(-12, -5)
ax3.set_ylabel('height / km')
ax3.set_xlabel('density')
ax3.set_title('North Hemisphere fall 0')
ax3.set_yticks([40,60,80,100, 120, 140])
ax3.grid(True)

ax4.set_ylim(40, 150)
ax4.set_xlim(-12, -5)
ax4.set_ylabel('height / km')
ax4.set_xlabel('density')
ax4.set_title('North Hemisphere fall 40 N')
ax4.set_yticks([40,60,80,100, 120, 140])
ax4.grid(True)

ax5.set_ylim(40, 150)
ax5.set_xlim(-12, -5)
ax5.set_ylabel('height / km')
ax5.set_xlabel('density')
ax5.set_title('North Hemisphere fall 80 N')
ax5.set_yticks([40,60,80,100, 120, 140])
ax5.grid(True)


ax6.set_visible(False)


#Extra stuff

plt.tight_layout()

fig.legend([data80s, regression80s, regression80s1 ], labels = ['data ', 'polyfit degree 6', 'polyfit degree {}'.format(b)], loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('fall_1d_regression_subplots.pdf', bbox_inches='tight')

```
